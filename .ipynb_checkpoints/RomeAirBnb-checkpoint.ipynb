## Data Analysis

### Main Research Question 1

#### 1. What are the pricing and popularity hotspots in Rome

```python
# Import libraries
import os
import pyproj
import matplotlib.pyplot as plt
```

```python
# Due to an issue on my laptop, this script determines where the database file (proj.db) is which maps the
# earth's coordinates. This is usually a conda to python connection issue.

# 1. Ask pyproj where the data directory is
path_to_proj_data = pyproj.datadir.get_data_dir()

# 2. Force the environment variable to use this path
os.environ['PROJ_LIB'] = path_to_proj_data

# 3. Now import geopandas
import geopandas as gpd

print(f"PROJ data found at: {path_to_proj_data}")

# Try loading your file
df_geo_rome = gpd.read_file("neighbourhoods.geojson")
```

```python
# Turn CSV into GeoDF
df_geo_listings = gpd.GeoDataFrame(df_listings,
                                   geometry=gpd.points_from_xy(df_listings["longitude"], df_listings["latitude"]),
                                   crs="EPSG:4326"
                                  )
```

```python
# Spatial Join (Points from CSV + Polygons of neighbourhoods

# Ensure both map layers use the same projection (Required!)
df_geo_listings = df_geo_listings.to_crs(df_geo_rome.crs)

# 2. Perform the Spatial Join
# "Find which neighbourhood every point belongs to"
joined = gpd.sjoin(df_geo_listings, df_geo_rome, how="inner", predicate="within")

# 3. Count the points per neighbourhood
# Assuming the neighbourhoods file has a column named 'neighbourhood_name'
counts = joined.groupby(by="neighbourhood_right").size().reset_index(name='point_count')

# 4. Merge the counts back into the original neighbourhood map
final_map = df_geo_rome.merge(counts, left_on="neighbourhood", right_on="neighbourhood_right", how='left')

# Fill NaN values with 0 (for neighbourhoods with no points)
final_map['point_count'] = final_map['point_count'].fillna(0)
```

```python
# Note this only answers chloroplot map of how many dots are per neighbourhood
final_map.plot(column='point_count', legend=True)
```
